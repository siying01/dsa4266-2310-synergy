```python
### Note: Uncompress json.gz file
### Note: Install packages

# load packages
import json
import pandas as pd
from tqdm import tqdm
import time

# Define file path for training data
dat_path = "C:/Users/joann/Downloads/NUS/Y4S1/DSA4266/Project2/Datasets/dataset0.json"

# Set up data frame
df = pd.DataFrame()
name, pos, seq = [], [], []
f1, f2, f3, f4, f5, f6, f7, f8, f9 = [], [], [], [], [], [], [], [], []

# Read and extract data
with open(dat_path) as file:
    for line in tqdm(file): # each line is for a specific transcript-position-6seq
        row = json.loads(line)

        for k1, v1 in row.items(): # each transcript id:{...}
            
            for k2, v2 in v1.items(): # each position:{...}
                
                for k3, v3 in v2.items(): # each 6seq:{list of reads}
                    
                    for ele in v3: # each read
                        name.append(k1)
                        pos.append(k2)
                        seq.append(k3)
                        
                        f1.append(ele[0])
                        f2.append(ele[1])
                        f3.append(ele[2])
                        f4.append(ele[3])
                        f5.append(ele[4])
                        f6.append(ele[5])
                        f7.append(ele[6])
                        f8.append(ele[7])
                        f9.append(ele[8])
                        
# Assign extracted data into dataframe

df["transcript_id"] = name
df["transcript_position"] = pos
df["6_seq"] = seq

df["f1"] = f1
df["f2"] = f2
df["f3"] = f3
df["f4"] = f4
df["f5"] = f5
df["f6"] = f6
df["f7"] = f7
df["f8"] = f8
df["f9"] = f9

# Check on dataframe
df.head()

# Define file path for labels and gene info
dat2_path = "C:/Users/joann/Downloads/NUS/Y4S1/DSA4266/Project2/Datasets/data.info"

# Read data
dat2 = pd.read_csv(dat2_path)

# Check data
dat2.head()

# Check data dtypes
dat2.dtypes
df.dtypes

# Ensure correct dtypes for merging based on columns and for saving space
df['transcript_position']=df['transcript_position'].astype("int32")
dat2['transcript_position']=dat2['transcript_position'].astype("int32")
dat2['label']=dat2['label'].astype("int32")

# Merge training data with labels and gene info
result = pd.merge(df, dat2, on=["transcript_id", "transcript_position"])
result.head()

# Write data to csv
start_time = time.time()
result.to_csv("data_v1.csv")
end_time = time.time()
print(end_time - start_time)

### Note: Compress csv file
```
